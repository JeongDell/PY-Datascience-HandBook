### Matplotlib 주의사항

대부분의 plt함수가 ax메서드로 바꿀 수 있지만 모든 명령어에 해당하는 것은 아님
- 특히 축의 경계와, 레이블, 제목을 설정하는 함수는 약간 수정됨


매트랩 스타일 함수와 객체지햐 메서드 사이에서 전환할 때 다음과 같이 바꿔야 함

- plt.xlabel() -> ax.set_xlabel()

- plt.ylabel() -> ax.set_ylabel()

- plt.xlim() -> ax.set_xlim()

- plt.ylim() -> ax.set_ylim()

- plt.title() -> ax.set_title()

객체지향 인터페이스에서는 ax.set()를 사용해서 모든 속성을 한번에 설정할 수 있음

```python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim = (0, 10), ylim = (-2, 2), xlabel = 'x', ylabel = 'sin(x)', title = 'A Simple Plot')
```

```python
#정규분포 속 랜덤 난수로 x,y,색깔,크기 지정
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

#그림 크기 설정
plt.figure(figsize = (10, 7))

#alpha: 투명도 지정(0: 완전 투명. 1: 불투명)
plt.scatter(x, y, c = colors, s = sizes, alpha = 0.3, cmap = 'viridis')

#색상 척도 표시
plt.colorbar()
```

#### 연속오차

plt.plot과 plt.fill_between을 결합해서 사용

```python
#가우시안 회귀모델을 사용해서 시각화 
from sklearn.gaussian_process import GaussianProcessRegressor
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel as C

# 모델을 정의하고 일부 데이터를 그림
model = lambda x: x * np.sin(x)
xdata = np.array([1, 3, 5, 6, 8])
ydata = model(xdata)

# 가우시안 프로세스 적합 계산
# 커널 정의: 상수 커널과 RBF 커널을 사용
kernel = C(1.0, (1e-3, 1e3)) * RBF(1, (1e-2, 1e2))
gp = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=10)

# 모델 적합
gp.fit(xdata[:, np.newaxis], ydata)

# 예측 데이터 생성
xfit = np.linspace(0, 10, 1000)
yfit, y_std = gp.predict(xfit[:, np.newaxis], return_std=True)

# 오차 범위 계산 (95% 신뢰 구간)
dyfit = 2 * y_std
```

```python
#결과 시각화

plt.plot(xdata,ydata,'or')
plt.plot(xfit,yfit,'-',color='gray')
```

```python
plt.fill_between(xfit,yfit-dyfit,yfit+dyfit,
                 color='gray',alpha=0.2)
plt.xlim(0,10)
```

```python
plt.plot(xdata,ydata,'or')
plt.plot(xfit,yfit,'-',color='gray')
plt.fill_between(xfit,yfit-dyfit,yfit+dyfit,
                 color='gray',alpha=0.2)
plt.xlim(0,10)
```

### 3차원 함수 시각화하기

```python
def f(x,y):
    return np.sin(x)**10 +np.cos(10+x*y)*np.cos(X)
```

```python
x = np.linspace(0,5,50)
y = np.linspace(0,5,40)

#meshgrid(): 1차원 배열로부터 2차원 그리드 생성
#X는 열방향으로 y만큼, 3만큼 확장
#Y는 행 방향으로 x만큼 확장
X,Y = np.meshgrid(x,y)
Z = f(X,Y)
```

```python
#음수 값은 점선으로, 양수값은 실선으로 표시
plt.figure(figsize= (8,6))
plt.contour(X,Y,Z, colors='black')
```

```python
plt.figure(figsize = (8, 6))
plt.contour(X, Y, Z, 20, cmap = 'RdGy')
```

```python
plt.figure(figsize = (8, 6))
plt.contourf(X, Y, Z, 20, cmap = 'RdGy')
plt.colorbar()
```

```python
plt.figure(figsize = (8, 6))
plt.imshow(Z, extent = [0, 5, 0, 5], origin = 'lower', cmap = 'RdGy')
plt.colorbar()
plt.axis(aspect = 'image')
```

```python
plt.figure(figsize = (8, 6))
contours = plt.contour(X, Y, Z, 3, colors = 'black')
plt.clabel(contours, inline = True, fontsize = 8)
plt.imshow(Z, extent = [0, 5, 0, 5], origin = 'lower', cmap = 'RdGy', alpha = 0.5)
plt.colorbar()
```

### 색상 지도 선택하기

순차적 색상 지도: 하나의 연속적인 순서를 가지는 색상으로 구성

발산적 색상 지도: 일반적으로 평균을 기즌으로 양수 및 음수 편차를 나타내는 두 가지 구별된 색상을 초함함

정성적 색상 지도: 특정 순서 없이 색을 혼합

```python
from matplotlib.colors import LinearSegmentedColormap

def grayscale_cmap(cmap):
    ''' 색상 지도의 회색조 버전을 반환'''
    cmap = plt.cm.get_cmap(cmap)
    colors = cmap(np.arange(cmap.N))
    
    # RGBA를 감지할 수 있는 회색조 밝기로 전환
    RGB_weight = [0.299, 0.587, 0.114]
    luminance = np.sqrt(np.dot(colors[:, :3] ** 2, RGB_weight))
    colors[:, :3] = luminance[:, np.newaxis]
    
    return LinearSegmentedColormap.from_list(cmap.name + "_gray", colors, cmap.N)

def view_colormap(cmap):
    ''' 색상 지도를 그에 상응하는 회색조 버전과 함께 플로팅'''
    cmap = plt.cm.get_cmap(cmap)
    colors = cmap(np.arange(cmap.N))
    cmap = grayscale_cmap(cmap)
    grayscale = cmap(np.arange(cmap.N))
    
    fig, ax = plt.subplots(2, figsize = (6, 2), subplot_kw = dict(xticks = [], yticks = []))
    ax[0].imshow([colors], extent = [0, 10, 0, 1])
    ax[1].imshow([grayscale], extent = [0, 10, 0, 1])
```

```python
view_colormap('jet')
```

```python
view_colormap('viridis')
```

```python
view_colormap('cubehelix')
```

```python
view_colormap('RdBu')
```
